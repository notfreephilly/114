````python
class DoublyLinkedList:
    class __Node:
        def __init__(self, data):
            self.data = data
            self.next = None
            self.prev = None

    def __init__(self):
        self.head = None
        self.tail = None
        self.size = 0

    def append(self, value):
        # Create new node instantiate then assign
        self.size += 1
        new_node = self.__Node(value)
        # check to see if list is empty
        if self.is_empty:
            self.head = new_node
            self.tail = new_node
        else:
           # current tail node point to new node
            self.tail.next = new_node
            # for  new node set previous pointer to current tail
            new_node.prev = self.tail
            # update tail the new node
            self.tail = new_node
            

    def insert(self, index, value):
        new_node = self.__Node(value)
        if index <= 0:
            self.head.prev = new_node
            new_node.next = self.head
            self.head = new_node
            # if the given index is bigger than the size of the list then append the value
        elif index >= self.size:
            self.append(new_node)

        else:
            current = self.head
            for _ in range(index-1):
                current = current.next
            new_node.next = current.next
            current.prev = new_node
            current.next = current
        self.size += 1

    def remove(self, value):
        # Remove the first instance of "value" in the list
        current = self.head
        prev = None
        found = False
        while current and not found:
            if current.data == value:
                found = True
            else:
                prev = current
                current = current.next
        if found:
            if not prev:
                self.head = self.head.next
            else:
                if current == self.tail:
                    self.tail = self.tail.prev
                next_node = current.next
                prev.next = next_node
                if next_node:
                    next_node.prev = prev
            self.size -= 1
        else:
            raise ValueError("Value not found")

    def index(self, value):
        pass

    def search(self, index):
        pass

    def __str__(self):
        out = "["
        current = self.head
        if current:
            out += "%s" % current.data
            current = current.next
            while current:
                out += ", %s" % current.data
                current = current.next
        out += "]"
        return out
        
    @property
    def is_empty(self):
        return self.head == None

    def __getitem__(self, index):
        return self.search(index)

    # try to implement this method
    def __setitem__(self, index):
        pass
````
